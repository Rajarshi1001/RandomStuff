```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from scipy.signal import find_peaks
%matplotlib inline
data = pd.read_csv('XRD.csv',usecols = range(2))
```

```python
data.head(5)
```

```python
X = data.iloc[:,0]
X = np.array(X)
y = data.iloc[:,1]
y = np.array(y)
fig = plt.figure(figsize=(30,10))
ax = fig.add_subplot(111)
max_intensity = np.max(y)
max_diffraction_angle = X[np.argmax(y)]
ax.plot(X,y,marker = 'o',linewidth = 2, linestyle = 'dashed',color='blue')
ax.set_ylabel("Intensity of the Diffracted Beam (Counts)",fontsize = 20)
ax.set_xlabel("Diffracting Angle (2 theta)", fontsize=20)
ax.plot(max_diffraction_angle,max_intensity,marker='o', markersize=20, markeredgecolor="red", markerfacecolor="black")
ax.annotate('Peak', xy=(max_diffraction_angle,max_intensity), xytext=(55, 49000),
            arrowprops=dict(facecolor='black', shrink=0.05),fontsize=20)
plt.grid()
plt.show()
print("Bragg's Angle (degrees): {}".format(max_diffraction_angle))
print("Max Intensity of Diffracted beam(Counts): {}".format(max_intensity))
```
